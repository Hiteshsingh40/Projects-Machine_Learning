## Model training

```python
import numpy as np
import pandas as pd
import sklearn
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
# Now Lets seprate imnput and output variable to be used for the linear clasification problem
x=df_salary.drop('Salary',axis=1)
x
```

```python
y=df_salary['Salary']
y
```

```python
print(x.shape,y.shape)
```

```python
# Scaling down using StandardScaler
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.tree import DecisionTreeRegressor
```

```python
#Creating a function

def max_r2_score(reg,x,y):
    max_r_score=0
    for r_state in range(42,101):
        x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=r_state,test_size=.20)
        reg.fit(x_train,y_train)
        y_pred=reg.predict(x_test)
        r2_scr=r2_score(y_test,y_pred)
        print("r2 score corresponding to",r_state,"is:",r2_scr)
        if r2_scr>max_r_score:
            max_r_score=r2_scr
            final_r_state=r_state
    print("max r2 score corresponding to",final_r_state,"is:",max_r_score)
    
    return final_r_state
```

```python
#Lets use Linear Regression

lreg=LinearRegression()
r_state=max_r2_score(lreg,x,y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=71,test_size=.20)
```

```python
print(x_train.shape,x_test.shape)
```

```python
#Using Linear Regression
lr=LinearRegression()
lr.fit(x_train,y_train)
print("Score:",lr.score(x_train,y_train))
print('\n')
print("Coefficient:",lr.coef_)
print('\n')
print("Intercept:",lr.intercept_)
print('\n')
predlr=lr.predict(x_test)
print("Predicted value:",predlr)
print("\n")
print("Actual value:",(y_test))
print('\n')
print('Mean absolute error:',mean_absolute_error(y_test,predlr))
print('Mean squared error:',mean_squared_error(y_test,predlr))
print('Root Mean Squared Error:',np.sqrt(mean_squared_error(y_test,predlr)))
print('\n')
print("r2_score:",r2_score(y_test,predlr))
```

```python
#Using algorithms via for loop

model=[LinearRegression(),Lasso(),Ridge(),ElasticNet(),DecisionTreeRegressor()]
for m in model:
    m.fit(x_train,y_train)
    print('Score of',m,'is:',m.score(x_train,y_train))
    predm=m.predict(x_test)
    print("Predicted Value:",predm)
    print('Error of',m,'is:')
    print('Mean absolute error:',mean_absolute_error(y_test,predm))
    print('Mean squared error:',mean_squared_error(y_test,predm))
    print('Root Mean Squared Error:',np.sqrt(mean_squared_error(y_test,predm)))
    print("r2_score:",r2_score(y_test,predm))
    print('************************************************************************************************')
    print('\n')
```

```python
#Cross Validate the model

model=[LinearRegression(),Lasso(),Ridge(),ElasticNet(),DecisionTreeRegressor()]
for m in model:
    score=cross_val_score(m,x,y,cv=5,scoring='r2')
    print('Score of',m,'is:')
    print("Score:",score)
    print('Mean Score:',score.mean())
    print("Standard deviation:",score.std())
    print('************************************************************************************************')
    print('\n')
```

```python
#Using Random Forest

from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(n_estimators=9,random_state=71)
rf.fit(x_train,y_train)
print('Score:',rf.score(x_train,y_train))
predrf=rf.predict(x_test)
print('\n')
print('Mean absolute error:',mean_absolute_error(y_test,predrf))
print('Mean squared error:',mean_squared_error(y_test,predrf))
print('Root Mean Squared Error:',np.sqrt(mean_squared_error(y_test,predrf)))
print('\n')
print("r2_score:",r2_score(y_test,predlr))
```

```python
#Using Ada Boost

from sklearn.ensemble import AdaBoostRegressor
adb=AdaBoostRegressor(n_estimators=350,learning_rate=0.001)
adb.fit(x_train,y_train)
print('Score:',adb.score(x_train,y_train))
predadb=adb.predict(x_test)
print('\n')
print('Mean absolute error:',mean_absolute_error(y_test,predadb))
print('Mean squared error:',mean_squared_error(y_test,predadb))
print('Root Mean Squared Error:',np.sqrt(mean_squared_error(y_test,predadb)))
print('\n')
print("r2_score:",r2_score(y_test,predadb))
```

```python
# Coross Valiodate RandomForestRegressor

score=cross_val_score(rf,x,y,cv=5,scoring='r2')
print("Score:",score)
print('Mean Score:',score.mean())
print("Standard deviation:",score.std())
```

```python
# Coross Valiodate AdaBoostRegressor

score=cross_val_score(adb,x,y,cv=5,scoring='r2')
print("Score:",score)
print('Mean Score:',score.mean())
print("Standard deviation:",score.std())
print("r2_score:",r2_score(y_test,predadb))
```
