```python
#The data is categorial so I convert it with LabelEncoder

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for col in df.columns:
    df[col] = le.fit_transform(df[col])
 
df.head()
```

```python
#From the table above it can be seen that the column "veil-type" is 0 and not contributing to the data so I remove it.

df=df.drop(["veil-type"],axis=1)
print(df.shape)
df.head()
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(df.corr(),annot=True)
```

```python
df_x=df.iloc[:,1:22]
df_x.head()
```

```python
df_y=df.iloc[:,0]
df_y.head()
```

```python
# import library
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression 
from sklearn.naive_bayes import GaussianNB      
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier


from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

from sklearn.model_selection import cross_val_score
```

```python
# Here we will findout random state value with which our Classification model leans maximum

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from sklearn import linear_model
max_r_score=0
for r_state in range(42,101):
    x_train,x_test,y_train,y_test=train_test_split(df_x,df_y,random_state=r_state,test_size=0.30)
    regr= linear_model.LogisticRegression()
    regr.fit(x_train,y_train)
    y_pred= regr.predict(x_test)
    r2_scr=r2_score(y_test,y_pred)
    if r2_scr>max_r_score:
        max_r_score=r2_scr
        final_r_state=r_state
print()
print()
print('max r2 score corresponding to',final_r_state,'is ',max_r_score)
```

```python
print(x_train.shape,x_test.shape)
```

```python
print(y_train.shape,y_test.shape)
```

```python
x_train,x_test,y_train,y_test=train_test_split(df_x,df_y,random_state=47,test_size=.30)
```

```python
#Applying Algorithms

model=[LogisticRegression(),GaussianNB(),DecisionTreeClassifier(),KNeighborsClassifier(),SVC()]

for m in model:
    m.fit(x_train,y_train)
    m.score(x_train,y_train)
    predm=m.predict(x_test)
    print('Accuracy score of',m,'is:')
    print("Accuracy Score:",accuracy_score(y_test,predm))
    print(confusion_matrix(y_test,predm))
    print(classification_report(y_test,predm))
    print('-------------------------------------------------------------')
    print('\n')
```

```python
# Above all Decision Tree Classifier is working well
#DecisionTreeClassifier(criterion='gini')------------->default
#Two criterion ginni and entropy

dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
dtc.score(x_train,y_train)
preddtc=dtc.predict(x_test)
print(accuracy_score(y_test,preddtc))
print(confusion_matrix(y_test,preddtc))
print(classification_report(y_test,preddtc))
```

```python
#Printing roc curve

from sklearn.metrics import roc_curve,roc_auc_score

y_pred_prob=dtc.predict_proba(x_test)[:,0]
fpr,tpr,thresholds=roc_curve(y_test,y_pred_prob)
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr,tpr,label='Decision tree Classifier')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title("Decision Tree Classifier")
plt.show()
```

```python
#Printing auc_score
auc_score=roc_auc_score(y_test,dtc.predict(x_test))
auc_score
```

```python
#Using Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier

rf=RandomForestClassifier(criterion='entropy', random_state=47,n_estimators=100)
rf.fit(x_train,y_train)
predrf=rf.predict(x_test)
print(accuracy_score(y_test,predrf))
print(confusion_matrix(y_test,predrf))
print(classification_report(y_test,predrf))
```

```python
# Cross Validated all algorithms by using for loop

model=[GaussianNB(),LogisticRegression(),SVC(),DecisionTreeClassifier(),KNeighborsClassifier()]


for m in model:
    score=cross_val_score(m,df_x,df_y,cv=10)
    print('Score of',m,'is:')
    print(score)
    print('Mean:',score.mean())
    print('Std:',score.std())
    print('*************************************************************************************************')
    print('\n')
```

```python
#Here Random Forest and Decision Tree are working well.
#So cross validate both models to see overfitting

from sklearn.model_selection import cross_val_score
```

```python
#Cross validate Decision Tree

score=cross_val_score(dtc,df_x,df_y,cv=10,scoring='accuracy')
print(score)
print('Mean:',score.mean())
print('Std:',score.std())
```

```python
#Cross validate Decision Tree

score=cross_val_score(rf,df_x,df_y,cv=10,scoring='accuracy')
print(score)
print('Mean:',score.mean())
print('Std:',score.std())
```
